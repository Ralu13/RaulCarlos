```python
import pandas as pd
import matplotlib.pyplot as plt
url= 'https://raw.githubusercontent.com/ehinar2112/MiDataSetEjemplo1/main/uso_memoria.txt'
```

```python
try:
    df=pd.read_csv(url)
    print("Primeras filas del dataset")
    print(df.head())
except Exception as e:
    print("Error al leer el dataset:", e)
```

```python
columna_de_interes = ' Uso de Memoria (%)'
plt.hist(df[columna_de_interes], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de '+columna_de_interes)
plt.show()
```

```python
x= 'Marca de Tiempo'
y= ' Uso de Memoria (%)'
plt.plot(df[x], df[y], label='Serie 1')
plt.xlabel('Tiempo')
plt.ylabel('Y')
plt.title('Grafico de Dispersion')
plt.show()
```
